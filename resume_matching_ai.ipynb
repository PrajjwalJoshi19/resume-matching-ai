```python
import nltk
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
```

```python
def preprocess(text):
    text = text.lower()
    text = "".join([char for char in text if char not in string.punctuation])
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)
```

```python
print("Paste Resume Text Below:")
resume = input()

print("\nPaste Job Description Below:")
job_description = input()
```

```python
resume_clean = preprocess(resume)
job_clean = preprocess(job_description)

print("Cleaned Resume:", resume_clean)
print("\nCleaned Job Description:", job_clean)
```

```python
vectorizer = TfidfVectorizer()

vectors = vectorizer.fit_transform([resume_clean, job_clean])

similarity = cosine_similarity(vectors[0:1], vectors[1:2])

score = similarity[0][0] * 100

print(f"\nResume Match Score: {score:.2f}%")
```
